```python
numeric_cols = df.select_dtypes(include=["int", "float"]).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap= "coolwarm", linewidths=5)
plt.title("Correlation Matrix Heatmap")
```

```python
position_counts = df["Position"].value_counts()
position_counts
```

```python
plt.pie(position_counts, labels=position_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
avg_attack_by_country = df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False).tail(5)
```

```python
avg_attack_by_country.sort_values(ascending=False).plot(kind = "bar") #all
avg_attack_by_country.sort_values(ascending=False).head(5).plot(kind = "bar") # first 5
plt.title("Average Attack by Country")
plt.xlabel("Country")
plt.ylabel("Attack")
```

```python
avg_serve_by_age = df.groupby("Age")["Serve"].mean()
avg_serve_by_age.sort_values(ascending=False)
```

```python
country_pos_attack = df.groupby(["Country", "Position"])["Attack"].max().reset_index()
country_pos_attack.sort_values(ascending=False, by="Attack").head(20)
```

```python
df.groupby("Country")["Dig"].sum()
```

```python
df.rename(columns={"Receive" : "Received"}, inplace=True)
```

```python
plt.scatter(df["Block"], df["Received"])
plt.title("Scatter Plot : Block vs Receive")
plt.xlabel("Block")
plt.ylabel("Received")
plt.show()
```

```python
sns.boxplot(x = df["Serve"])
plt.title("Box Plot : Distribution Of Serve Values")
```

```python
plt.hist(df["Age"], bins = 20, color = "orange", edgecolor = "black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
avg_attack_by_position = df.groupby("Position")["Attack"].mean()
avg_attack_by_position
avg_attack_by_position.sort_values(ascending=False).plot(kind = "bar", color = "violet", edgecolor = "purple")
plt.title("Average Attack by Position")
plt.xlabel("Position")
plt.ylabel("Attack")
```

```python
serve_trend_by_age = df.groupby("Age")["Serve"].mean()
serve_trend_by_age
serve_trend_by_age.plot(kind = "line", marker = "o", linestyle = "-", color="orange")
plt.title("Serve Trend Over Age Group")
plt.xlabel("Age")
plt.ylabel("Average Server")
```

```python
total_attack_block_by_country  = df.groupby("Country")[["Attack", "Block"]].sum()
total_attack_block_by_country
total_attack_block_by_country.sort_values(ascending=True, by="Attack").plot(kind="bar", stacked=True, colormap="viridis")
plt.title("Total Attack and Block by Country")
plt.xlabel("Country")
plt.ylabel("Attack")
```
